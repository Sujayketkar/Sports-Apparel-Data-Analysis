## <b><span style='color:#df2c14'> </span><span style='color:#ff6242'>Sales Performance Analysis: Which products are excelling in sales, and which are underperforming?</span></b>

```python
Product_Sales = df.groupby(by='Product')['Total Sales'].sum().reset_index(name='Total_Sales').sort_values(by='Total_Sales', ascending=False)
```

```python
Product_Sales
```

Sales analysis tells us that last 3 products may be under performing, however we will check units sold and operating profit to see what the data tells us

```python
Product_Sold = df.groupby(by='Product')['Units Sold'].sum().reset_index(name='Units_Sold').sort_values(by='Units_Sold', ascending=False)
```

```python
Product_Sold
```

```python
Product_OP = df.groupby(by='Product')['Operating Profit'].sum().reset_index(name='Operating Profit').sort_values(by='Operating Profit', ascending=False)
```

```python
Product_OP
```

## <b><span style='color:#df2c14'> </span><span style='color:#ff6242'>Men's Street Footwear,Women's Apparel,Men's Athletic Footwear are the products that are doing well. Men's Street Footwear bringing $208,826,244 in Sales.   	</span></b>

## <b><span style='color:#df2c14'> </span><span style='color:#ff6242'>Women's Street Footwear,Women's Athletic Footwear,Men's Apparel are the products that have opportunity to improvise their presence as can be seen by units sold and Operating profit data	</span></b>
